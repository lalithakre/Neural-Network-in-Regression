```python
import pickle
```

```python
from sklearn.datasets import load_diabetes
```

```python
obj=load_diabetes()
```

```python
x=obj.data
y=obj.target
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x1=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x1,y,test_size=0.2)
```

```python
from keras.models import Sequential 
sq=Sequential()
from keras.layers import Dense
#d=Dense
sq.add(Dense(12,activation='relu',input_dim=10))
sq.add(Dense(11,activation='relu'))
sq.add(Dense(10,activation='relu'))
sq.add(Dense(1,activation='relu'))
sq.compile(optimizer='adam',loss='mean_squared_error')
```

```python
sq.fit(xtrain,ytrain,epochs=200)
```

```python
ypred=sq.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```
